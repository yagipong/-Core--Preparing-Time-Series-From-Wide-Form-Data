```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.ticker as mticks
```

```python
url = 'Data/city.csv'
df = pd.read_csv(url)
df.head()
```

```python
largest_cities_df = df[df['SizeRank']<=3]
largest_cities_df
```

```python
melted = pd.melt(largest_cities_df ,id_vars = ['RegionID', 'SizeRank', 'RegionName',
                                      'RegionType', 'StateName', 'State',
                                      'Metro', 'CountyName'],
                                       value_name='HouseValue', var_name="Date")
melted.head()
```

```python
melted['Date'] = pd.to_datetime(melted['Date'])

melted
```

```python
melted = melted.set_index("Date")

melted
```

```python
ts = melted.groupby('RegionName').resample('M').first()

ts
```

```python
unstacked = ts.unstack(level=0)

unstacked
```

```python
def thousands(x, pos):
    new_x = x/1_000
    return f"{new_x:,.0f}K"

thousands_formatter = mticks.FuncFormatter(thousands)
```

```python
ax = unstacked['HouseValue'].plot()

ax.legend(loc = 'upper right',bbox_to_anchor=(1.5, 1))

ax.yaxis.set_major_formatter(thousands_formatter)
```

```python
unstacked_house_value = unstacked['HouseValue']

unstacked_diff = unstacked_house_value.diff()

unstacked_diff = unstacked_diff.loc['11/2008':'12/2008']

unstacked_diff
```
